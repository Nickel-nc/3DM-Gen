```python
from video_generator import VideoGenerator, extract_beats
```

```python
start = [
    [45, -20, 20],   # 'rightMiddle'
    [45, 110, 20],   # 'rightFront'
    [-45, 110, 20],  # 'leftFront'
    [-45, -20, 20],  # 'leftMiddle'
    [0, -20, 20],    # 'leftBack'
    [0, -20, 20]     # 'rightBack'
]
end = [
    [45, -20, 20],
    [0, 110, -80],
    [-90, 110, -80],
    [-45, -20, 20],
    [0, -20, 20],
    [0, -20, 20]
]
new_end = [
    [45, -20, 20],
    [90, 110, -80],
    [0, 110, -80],
    [-45, -20, 20],
    [0, -20, 20],
    [0, -20, 20]
]
```

```python
def moves_gen(beats):
    moves_l = [start, new_end]
    moves = []
    for ix, b in enumerate(beats):
        moves.append(moves_l[ix%2])
    return moves
```

```python
vgen = VideoGenerator(extract_beats, moves_gen)
```
